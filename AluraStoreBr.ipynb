```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

```python
import matplotlib.dates as mdates

# Criando a figura e os subgráficos (2x2)
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

# Pegando todas as lojas únicas
lojas = historico_faturamento_loja['Origem'].unique()
categorias = historico_faturamento_loja['Categoria do Produto'].unique()

# Obter a paleta de cores automaticamente
mapa_cores = colormaps.get_cmap('tab10')  # Paleta automática

# Criar um dicionário de cores para cada categoria
cores_dict = {categoria: mapa_cores(i / len(categorias)) for i, categoria in enumerate(categorias)}

# Iterar sobre as lojas e gerar os gráficos
for i, loja in enumerate(lojas):
    ax = axes[i // 2, i % 2]  # Definir a posição no grid 2x2
    df_loja_graf_4 = historico_faturamento_loja[historico_faturamento_loja['Origem'] == loja]  # Filtrar os dados da loja

    # Iterar sobre categorias e desenhar linhas
    for categoria in categorias:
        subset = df_loja_graf_4[df_loja_graf_4['Categoria do Produto'] == categoria]
        ax.plot(subset['Ano_Mês'], subset['Preço'], label=categoria, color=cores_dict[categoria])

    ax.set_title(f'Faturamento - {loja}')
    ax.set_xlabel('Ano | Mês')
    ax.set_ylabel('Faturamento')
    ax.tick_params(axis='x', rotation=45)  # Rotate x-axis labels for better readability
    ax.xaxis.set_major_locator(mdates.DayLocator(interval=5))   
    ax.legend(title="Categoria do Produto")  # Adiciona legenda para as categorias

# Ajustar layout para evitar sobreposição
plt.tight_layout()
plt.show()
```

```python
# Criando a figura e os subgráficos (2x2)
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

# Pegando todas as lojas únicas
lojas = historico_faturamento_loja['Origem'].unique()
categorias = historico_faturamento_loja['Categoria do Produto'].unique()

# Obter a paleta de cores automaticamente
mapa_cores = colormaps.get_cmap('tab10')  # Paleta automática

# Criar um dicionário de cores para cada categoria
cores_dict = {categoria: mapa_cores(i / len(categorias)) for i, categoria in enumerate(categorias)}

# Iterar sobre as lojas e gerar os gráficos
for i, loja in enumerate(lojas):
    ax = axes[i // 2, i % 2]  # Definir a posição no grid 2x2
    df_loja_graf_4 = historico_faturamento_loja[historico_faturamento_loja['Origem'] == loja]  # Filtrar os dados da loja

    # Inicializar a base para empilhamento
    bottom = pd.Series(0, index=df_loja_graf_4['Ano_Mês'].unique())

    # Iterar sobre categorias e desenhar barras empilhadas
    for categoria in categorias:
        subset = df_loja_graf_4[df_loja_graf_4['Categoria do Produto'] == categoria]
        subset = subset.set_index('Ano_Mês')  # Garantir alinhamento por índice
        aligned_bottom = bottom.reindex(subset.index, fill_value=0)  # Alinhar com o índice do subset
        ax.bar(
            subset.index, 
            subset['Preço'], 
            label=categoria, 
            color=cores_dict[categoria], 
            bottom=aligned_bottom
        )
        # Atualizar a base para empilhamento
        bottom = bottom.add(subset['Preço'], fill_value=0)

    ax.set_title(f'Faturamento - {loja}')
    ax.set_xlabel('Ano | Mês')
    ax.set_ylabel('Faturamento')
    ax.tick_params(axis='x', rotation=45)  # Rotate x-axis labels for better readability
    ax.xaxis.set_major_locator(mdates.DayLocator(interval=3))  
    ax.legend(title="Categoria do Produto")  # Adiciona legenda para as categorias

# Ajustar layout para evitar sobreposição
plt.tight_layout()
plt.show()
```
